## Description of the dataset

In this part of the assignment we will cluster the observations in the 20 Newsgroups dataset using the k-means algorithm. Each row of the dataset represents a document with bag of words features. If we were not given the labels for each document (i.e. the newsgroup it came from), clustering could allow us to infer which documents should have the same label. Observing common words within each cluster may allow us to give meaning to these inferred labels too.

### Description of the Data

This dataset is a collection of approximately 20,000 newsgroup documents, partitioned (nearly) evenly across 20 different newsgroups, each corresponding to a different topic. Some of the newsgroups are very closely related to each other (e.g. comp.sys.ibm.pc.hardware, comp.sys.mac.hardware), while others are highly unrelated (e.g misc.forsale, soc.religion.christian). 

To save you time and to make the problem manageable with limited computational resources, we preprocessed the original dataset. We will use documents from only 5 out of the 20 newsgroups, which results in a 5-class problem. More specifically the 5 classes correspond to the following newsgroups: 
1. `alt.atheism`
2. `comp.sys.ibm.pc.hardware`
3. `comp.sys.mac.hardware`
4. `rec.sport.baseball`
5. `rec.sport.hockey `

However, note here that classes 2-3 and 4-5 are rather closely related.

**In contrast to Assignment 1**, we have opted to use tf-idf weights ([term frequency - inverse document frequency](https://en.wikipedia.org/wiki/Tf%E2%80%93idf)) for each word instead of the frequency counts (which is why we will be downloading it anew). These weights represent the importance of a word to a document with respect to a collection of documents. The importance increases proportionally to the number of times a word appears in the document and decreases proportionally to the number of times the word appears in the whole corpus. 

Additionally we preprocess the data to include the most frequent 1000 words that appear in more than two documents, less than half of all documents, and that are not [stop words](https://en.wikipedia.org/wiki/Stop_words). We have provided the code for this preprocessing for you.

(a) ***My answer:***

The adjusted Rand index is a function that measures the similarity of the ground truth class assignments and our clustering algorithm assignments, ignoring permutations and with chance normalization.

The ARI measures the quality of the fit by calculating the similarity of all pair-wise comparisons between clusterings specified by a random model. 

A disadvantage of ARI is that it requires knowledge of the ground truth classes, these often to be establised by knowledgable annotators, similar to supervised learning, and therefore rarely available, as this is an expensive, time-consuming process. 


````python
#(b) # Computing and displaying the Adjusted Rand Index of the fitted model:

y_pred = kmc.predict(X=X)
adj_rnd = adjusted_rand_score(labels_pred=y_pred, labels_true=y)

print("The adjusted rand score is {0:.3g} to 3 decimal places.".format(adj_rnd))
````

The value of the ARI ranges from 0.0 to 1.0, although if the index is less than the expected index it can yield negative values. When the ARI is close to 0.0 this suggests a random cluster-independent labelling. When the ARI is close to 1.0 this stands for a complete similarity match.

(a) ***My answer:***

The very high dimensional nature of many data sets makes the direct visualization needed to pick a subset of the features impossible as we humans can only comprehend three dimensions. PCA in these cases presents a solution being an unsupervised data dimension reduction technique. This well-established mathematical technique allows the reduction of the dimensionality of data without the need for human visualisation.



````python
#(b) # Fitting a Kmeans classifier with PCA reduced X for 10 values:

adj_rand_list = []    
n_vals = [3, 6, 15, 47, 54, 78, 142, 285, 678, 987]

# Looping over each value of n
for n in n_vals :
    
    # Transforming X with PCA
    log_n = int(np.log(n, dtype=float64))
    X_PCA = PCA(n_components=log_n, random_state=1000).fit_transform(X)
    
    # Calculating the k-means cluster predictions with the pca transformed X
    kmc_pca = KMeans(n_clusters=5, random_state=1000) 
    kmc_pca.fit(X_PCA, y)
    y_pred = kmc_pca.predict(X_PCA)
    
    #Calculating the adjusted rand score for the value and storing it in the list
    adj_rand_i = adjusted_rand_score(y, y_pred)
    adj_rand_list.append(adj_rand_i)

n_with_best_ari = n_vals[np.argmax(adj_rand_list)]
best_ari = np.amax(adj_rand_list)

print("The value of n = {0}, produced the best ARI of {1:.3g} (to 3 d.p)"
      .format(n_with_best_ari, best_ari))
````

````python
#(c) # Plotting the the adjusted_rand_score against the number of principal components:

# Setting the figure size
plt.figure(figsize=(8,5))

# Setting the log scal and plotting the ARI values
plt.xscale('log')
plt.scatter(n_vals, adj_rand_list)
plt.plot(n_vals, adj_rand_list)

# Setting the title and labels
plt.title('ARI vs the number of Principal Components')
plt.ylabel('ARI')
plt.xlabel('Number of Principal Components')

# Displaying the plot
plt.show()
````

(d) ***My answer:***

The dimensionality reduction has had a marginal improvement on the ARI(originally having a ARI of 0.226, now with best value 0.268). This suggests that the dimensions are largely independent of one another and reducing them has a limited effect on the clustering clarity of the classifier.

````python
#(e) # Fitting a 5-cluster K-Means object to the data with the dimensionality that produced the best ARI:

# Fitting a transformed X with the value n that had the best ari
X_PCA = PCA(n_components=n_with_best_ari, random_state=1000).fit_transform(X)

# Calculating the kmc with the best ari
kmc_best_ari = KMeans(n_clusters=5, random_state=1000)  
kmc_best_ari.fit(X_PCA, y)

# Calculating the amount of values in each cluster
cluster_freq= Counter(kmc_best_ari.labels_)

# Creating the labels for each cluster
labels = []
classes = ['alt.atheism', 'rec.sport.hockey',  'comp.sys.mac.hardware', 'rec.sport.baseball', 'comp.sys.ibm.pc.hardware']

for label in range(5):
    labels.append("Cluster {0} \n{1} \nTotal data-points: {2}".format(label, classes[label], cluster_freq[label]))

# Plotting the figure
plt.figure(figsize=(15,10))
sns.countplot(x=kmc_best_ari.labels_, hue=y)

# setting the title and the labels
plt.title("Count of data-points for each class based cluster centre")
plt.xlabel("Cluster centers")
plt.ylabel("Count of Data Points")
plt.xticks([0,1,2,3,4], labels)

# displaying the plot
plt.show()
````

The clusters generally have a marginally clearer lead than previously, this suggests that PCA has helped somewhat.
````python
#(b) # Your Code goes here:

# Plotting Voronoi method taken from the mentioned utils package
def plot_voronoi(predictor, _range, sampling=100, clr_map='Set1'):
    # Sort out the Parameters
    x_min, x_max, y_min, y_max = _range
    if hasattr(sampling, '__len__'):
        assert len(sampling) == 2, 'sampling must be of size 2 or a scalar'
        step_x = (x_max - x_min)/sampling[0]
        step_y = (y_max - y_min)/sampling[1]
    else:
        step_x = (x_max - x_min) / sampling
        step_y = (y_max - y_min) / sampling

    # Create Mesh and Predict on it
    xx, yy = np.meshgrid(np.arange(x_min, x_max, step_x), np.arange(y_min, y_max, step_y))
    zz = predictor.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

    # Create Color Map
    _N = zz.max() - zz.min() + 1
    cmap = plt.get_cmap(clr_map, _N)
    
    # Plot 
    return plt.imshow(zz, interpolation='nearest', extent=_range, aspect='auto', origin='lower', 
                      alpha=.5, cmap=cmap, vmin=-0.5, vmax=_N-0.5)

# Fitting a transformed X with 2 pca companants
X_PCA_two = PCA(n_components=2, random_state=1000).fit_transform(X)

# Calculating the kmc with the best ari
kmc_two_pca = KMeans(n_clusters=5, random_state=1000)  
kmc_two_pca.fit(X_PCA, y)


# storing the cluster centers
clust_centers = kmc_two_pca.cluster_centers_

# category values
colours = ['red', 'purple', 'gold', 'blue', 'green']
nums = [0, 1, 2, 3, 4]
labels = ['alt.atheism', 'rec.sport.hockey',  
          'comp.sys.mac.hardware', 'rec.sport.baseball', 'comp.sys.ibm.pc.hardware']

# Plotting the voronoi diagram
plt.figure(figsize=(7,5))
plot_voronoi(kmc_two_pca, [-0.4, 0.5, -0.4, 0.6], clr_map='Set2')

# plotting the values in each category
for cat_num in range(5):
    
    cat = np.array((y[:] == y[nums[cat_num]])*1).reshape(-1,1)*X_PCA_two
    cat = cat[~np.all(cat== 0, axis=1)]
    
    plt.scatter(cat[1:400, 0],cat[1:400, 1], alpha=0.2, color=colours[cat_num], label=labels[cat_num])
 
# plotting the cluster centers on the graph
plt.scatter(clust_centers[:, 0], clust_centers[:, 1], marker='x', s=200, color = 'k')

# displaying the title and legend
plt.title("Voronoi Diagram")
plt.legend()

# plotting the graph
plt.show()
````
